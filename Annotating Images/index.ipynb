# Adding Annotations to the image

Using OpenCV we will add rectangle boxes, circles or text to the image.

```python
# Import libraries

import cv2
from matplotlib import pyplot as plt 

# import numpy as np
```

#### Defining colors or annotations 

```python
yellow  = (0, 255, 255)
red     = (0, 0, 255)
magenta = (255, 0, 255)
green   = (0, 255, 0)
```

#### Reading the image in RGB

```python
image = cv2.imread("../content/dog.png")
plt.imshow(image)
```

#### Add red cirlces around eyes
Use ``circle()`` function.

```python
# Create the copy of the original image
image_circle=image.copy()
# Draw a red circle.
image_circle = cv2.circle(image_circle, (138, 108), 20, red, thickness = 2, lineType = cv2.LINE_AA)
image_circle = cv2.circle(image_circle, (235, 110), 20, red, thickness = 2, lineType = cv2.LINE_AA)

image_circle_rgb = cv2.cvtColor(image_circle, cv2.COLOR_BGR2RGB)
plt.imshow(image_circle_rgb)
```

#### About ``cv2.circle()``
``cv.circle()`` takes following params:
1. img: The image on which the circle will be drawn.

2. center: The coordinates of the center of the circle. It is a tuple (x, y) specifying the center's position, where (0, 0) is the top-left corner of the image.

3. radius: The radius of the circle.

4. color: The color of the circle. 

5. thickness (optional): The thickness of the circle's outline. If negative, the circle is filled. By default, the thickness is 1.

6. lineType (optional): Type of the circle boundary. It controls the line type. The default value is cv2.LINE_8, which specifies an 8-connected line. Other options include cv2.LINE_AA, which specifies an anti-aliased line.

#### Adding rectangle
Use ``cv.rectable()`` to draw rectangle in the image

```python
# Make a copy of the original image.
image_rectangle = image.copy()

# Draw a magenta rectangle.
image_rectangle = cv2.rectangle(image_rectangle, (160, 190), (215, 150), red, thickness = 2, lineType = cv2.LINE_8)

image_rectangle_rgb = cv2.cvtColor(image_rectangle, cv2.COLOR_BGR2RGB)
plt.imshow(image_rectangle_rgb)
```

#### Adding text
Use ``cv.putText()`` to write text in the image

```python
# Make a copy of the original image.
image_text = image.copy()

# Add text to the image.
text = 'My Dog'
font_face = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.5
font_color = red
font_thickness = 1

image_text = cv2.putText(image_text, text, (10, 20), font_face, font_scale, font_color, font_thickness, cv2.LINE_AA)

image_text_rgb = cv2.cvtColor(image_text, cv2.COLOR_BGR2RGB)

plt.imshow(image_text_rgb)
```
