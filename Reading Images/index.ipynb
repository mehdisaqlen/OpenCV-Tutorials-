```python
# Import libraries

import cv2
from matplotlib import pyplot as plt 
```

#### Read the image
We use ``imread()`` function from openCV to read the image it takes follwoing paramerters:
1. Filename (required) - path of the image to be shown
2. Flag (Optional)


```python
# Usage

image= cv2.imread("./dog.png")
```

### Parameters 
1. Filename
    - **Type** : ``str``
    - **Description**: Path to the image file.
    - **Example**: 'path_to_your_image.jpg'

2. Flags
    - **Type** : ``int``
    - **Description**: Specifies the color type of the loaded image.
    - **Default**: ``cv2.IMREAD_COLOR``
    - **Options**: 
        - ``cv2.IMREAD_COLOR`` (or 1): Loads a color image. Any transparency of image will be neglected. It is the default flag.
        - ``cv2.IMREAD_GRAYSCALE`` (or 0): Loads the image in grayscale mode.
        - ``cv2.IMREAD_UNCHANGED`` (or -1): Loads the image as is (including the alpha channel if present).

```python
plt.imshow(image)
# plt.axis('off')
```

We use ``matplotlib`` function ``imshow`` to display the image on the browser. You can hide the axis of the image by using ``plt.axis('off')``


#### Read the image in ``cv2.IMREAD_GRAYSCALE``

```python
image= cv2.imread("./dog.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(image)
```

### Why is the image not grey?
We have converted the image color to 'grayscale' using ``cv2.IMREAD_GRAYSCALE``. As we use ``plt.imshow()`` function to display the image we need to specify the *colormap* in ``imshow`` function. 

```python
plt.imshow(image, cmap='grey')
```

```python
image= cv2.imread("./dog.png")

# Convert the image from BGR to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_image)
```
